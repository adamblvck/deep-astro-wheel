```python
from skyfield.api import load
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from skyfield.api import utc
from scipy.optimize import brentq # machine learning

from datetime import timedelta, datetime
import pytz
```

```python
df = pd.read_csv('planets.csv', sep=';')
df.head(3)
```

```python
df[500:5000].plot()
```

```python
df_diff = pd.DataFrame()
df_diff['t'] = df['t']
df_diff['earth'] = df['earth'].diff()
df_diff['sun'] = df['sun'].diff()
df_diff['moon'] = df['moon'].diff()
df_diff['venus'] = df['mars'].diff()
df_diff['jupiter'] = df['jupiter'].diff()
```

```python
df_diff[500:5000].plot()
```

We need to remove those spikes somehow

```python
df_diff_cleaned_1 = pd.DataFrame()
planets = ['earth', 'sun', 'moon', 'venus', 'mars']
for p in planets:
    df_diff_cleaned_1[p] = df[p].diff()


df_diff_cleaned_1[ df_diff_cleaned_1 < -5 ] = 0
```

```python
df_diff_cleaned_1[500:1000].plot()
```

And now we peaks that reach to 0, instead of what it was... proposition to add +2pi where info goes beyond that

```python
df_diff_cleaned_2 = pd.DataFrame()
planets = ['earth', 'sun', 'venus', 'mars', 'jupiter'] # 'moon'
for p in planets:
    df_diff_cleaned_2[p] = df[p].diff()

    # add 2pi to levels where the difference this a fullcircle warp (we're substracting here)
    spike = df_diff_cleaned_2[p] < -5
    df_diff_cleaned_2[p][spike] = df_diff_cleaned_2[p][spike] + math.pi*2

    # add 2pi to levels where the difference this a fullcircle warp (we're adding here)
    spike = df_diff_cleaned_2[p] > 5
    df_diff_cleaned_2[p][spike] = df_diff_cleaned_2[p][spike] - math.pi*2


    # print (spike)
    # print (df_diff_cleaned_2[p][spike])


# df_diff_cleaned_2[ df_diff_cleaned_2 < -5 ] = offsetfix[ df_diff_cleaned_2 < -5 ]
```

```python
df_diff_cleaned_3 = pd.DataFrame()
planets = ['sun', 'jupiter'] # 'moon'
for p in planets:
    df_diff_cleaned_3[p] = df[p].diff()

    # add 2pi to levels where the difference this a fullcircle warp (we're substracting here)
    spike = df_diff_cleaned_3[p] < -5
    df_diff_cleaned_3[p][spike] = df_diff_cleaned_3[p][spike] + math.pi*2

    # add 2pi to levels where the difference this a fullcircle warp (we're adding here)
    spike = df_diff_cleaned_3[p] > 5
    df_diff_cleaned_3[p][spike] = df_diff_cleaned_3[p][spike] - math.pi*2
```
